```python
# --- Datos del Desafío ---
ciphertext = "label"
key_integer = 13

# Lista para almacenar los valores enteros resultantes del XOR
xor_integers = []

# --- 1. Aplicar XOR a cada carácter con la clave ---
for char in ciphertext:
    # Obtener el valor Unicode/ASCII del carácter (integer)
    char_ordinal = ord(char)

    # Aplicar la operación XOR (^) con la clave 13
    xor_result = char_ordinal ^ key_integer

    # Almacenar el resultado entero
    xor_integers.append(xor_result)

# --- 2. Convertir los enteros resultantes de vuelta a una cadena ---
new_string = ""
for integer in xor_integers:
    # Convertir el entero de vuelta a su carácter ASCII/Unicode
    new_string += chr(integer)

# --- 3. Formatear y Mostrar el Flag ---
# El desafío pide enviar el flag como crypto{new_string}
flag = f"crypto{{{new_string}}}"

print(f"Cadena Original: {ciphertext}")
print(f"Valores XOR (Enteros): {xor_integers}")
print(f"El Flag es: {flag}")
```

```python
def xor_bytes(a, b):
    # Función para aplicar XOR byte a byte
    return bytes([x ^ y for x, y in zip(a, b)])

# Claves de entrada (las variables proporcionadas)
KEY1 = bytes.fromhex("a6c8b6733c9b22de7bc0253266a3867df55acde8635e19c73313")
KEY2_xor_KEY1 = bytes.fromhex("37dcb292030faa90d07eec17e3b1c6d8daf94c35d4c9191a5e1e") # No se usa directamente en el cálculo final
KEY2_xor_KEY3 = bytes.fromhex("c1545756687e7573db23aa1c3452a098b71a7fbf0fddddde5fc1")
FINAL = bytes.fromhex("04ee9855208a2cd59091d04767ae47963170d1660df7f56f5faf")

# Lógica simplificada: FLAG = FINAL XOR KEY1 XOR KEY2_xor_KEY3
# Esto es mucho más directo y evita el cálculo de KEY2 y KEY3.
# Se reduce de 4 operaciones XOR a solo 2.
TEMP = xor_bytes(FINAL, KEY1)
FLAG = xor_bytes(TEMP, KEY2_xor_KEY3)

print(FLAG)
print("crypto{" + FLAG.decode() + "}")
```

```python
def xor_data(data, key_byte):
    """Aplica la operacion XOR a cada byte de 'data' con 'key_byte'."""
    # Repite el byte de la clave para hacer XOR con cada byte de los datos
    return bytes([b ^ key_byte for b in data])

# 1. Datos cifrados proporcionados en el desafío (hexadecimal)
HEX_CIPHERTEXT = "73626960647f6b206821204f21254f7d694f7624662065622127234f726927756d"

# 2. Decodificar el hex a bytes binarios reales
CIPHERTEXT = bytes.fromhex(HEX_CIPHERTEXT)

print(f"Longitud del mensaje cifrado (en bytes): {len(CIPHERTEXT)}")
print("--- Iniciando búsqueda de clave por fuerza bruta (0-255) ---")

# 3. Iterar sobre todas las 256 posibles claves de 1 byte (0 a 255)
for key in range(256):
    # Intentar descifrar el texto
    decrypted_bytes = xor_data(CIPHERTEXT, key)

    # 4. Intentar decodificar a texto. Usamos 'latin-1' porque mapea todos
    # los bytes (0-255) a caracteres, evitando errores de decodificación.
    decrypted_text = decrypted_bytes.decode('latin-1')

    # 5. Filtrar e imprimir el resultado que contiene la palabra clave 'crypto'
    if "crypto" in decrypted_text:
        print("\n✅ ¡CLAVE ENCONTRADA! ✅")
        print(f"Clave (byte decimal): {key}")
        print(f"Flag (Texto Descifrado): {decrypted_text}")

        # Una vez que encontramos la flag, salimos del bucle
        break

print("\n--- Búsqueda finalizada ---")
```
